# Hodgkin–Huxley


The voltage in the Hodgkin Huxley model of the action potential evolves according to the following differential equations:

\begin{equation}
\begin{aligned}
C \dot v &= 
g_\ell (E_\ell - v)
+ \bar g_\mathrm{K} n^4 (E_\mathrm{K}  - v) 
+ \bar g_\mathrm{Na}  m^3 h (E_\mathrm{Na} - v) 
+ I
\\
\dot n&= \alpha_{\mathrlap{n}\phantom{m}}(v)\cdot(1-\mathrlap{n}\phantom{m}) - \beta_{\mathrlap{n}\phantom{m}}(v)\cdot n
\\
\dot m&= \alpha_{m}(v)\cdot(1-m) - \beta_m(v)\cdot m
\\
\dot h&= \alpha_{\mathrlap{h}\phantom{m}}(v)\cdot(1-\mathrlap{h}\phantom{m}) - \beta_{\mathrlap{h}\phantom{m}}(v)\cdot h.
\end{aligned}
\end{equation}

## Python code for the gating variables $n$, $m$, $h$:

We will use [parameters for and excitatory pyramidal cell in rat cortex, taken from 
Table 2.1 in Gerstner et al. (2004)](https://neuronaldynamics.epfl.ch/online/Ch2.S2.html) ultimately derived from Mainen et al. (1995),
Huguenard et al. (1988), and
Hamill et al. (1991).


> ![](table.svg)
> > 
> **Table 2.1** from Gerstner et al. (2004) defining the  $\alpha(\cdot)$ and $\beta(\cdot)$ for the gating variables $n$, $m$, and $h$. 

These variables have a singularity at the threshold that we can remove using
L'Hôpital's rule:

\begin{equation}
\begin{aligned}
\lim_{x\to 0} \frac
{x}
{1 - e^{-x/r}}
=
\lim_{x\to 0} \frac
{\tfrac{\mathrm{d}}{\mathrm{d}x}x}
{\tfrac{\mathrm{d}}{\mathrm{d}x}(1 - e^{-x/r})}
=
\lim_{x\to 0} 
\frac
{1}
{\tfrac 1 c e^{x/r}}
=
\frac
{r}
{e^{0}}
=
r
\end{aligned}
\end{equation}

We defined a singularity-free piecewise helper function for $\alpha(\cdot)$, $\beta(\cdot)$:

\begin{equation}
\begin{aligned}
\operatorname{gate\_nosingularity}(d;r) &= \begin{cases}
\frac{x}{1-\exp(-x/r)} & x\ne 0 \\
r & x=0 \\
\end{cases}
\end{aligned}
\end{equation}

To integrate, we assume $\alpha(\cdot)$, $\beta(\cdot)$ constant over a small $\Delta t$.

It will be useful to write the ODE for gating variable $p\in\{n,m,h\}$ as
 
$$
\tau_p \dot p = p_\infty - p
$$

where

$$
\tau_p = \frac{1}{\alpha + \beta}
$$

and

$$
p_\infty = \frac{\alpha}{\alpha + \beta} = \alpha \tau.
$$
 
![](hhgateCA3.svg)

```python
gNmax = 40  # Sodium (Na⁺) maximal conductance (microsiemens)
gKmax = 35  # Potassium (K⁺) maximal conductance (microsiemens)
gL    = 0.3 # Leak conductance (microsiemens)

sexp = lambda x:exp(clip(x,-15,15)) # Exp wrapped to ignore overflow

def gate_nosingularity(x,r):
    return piecewise(x,[x!=0],[lambda x:x/(1-sexp(-x/r)), lambda x:r])

an   = lambda v: 0.020*gate_nosingularity(  v-25 ,9 )
bn   = lambda v: 0.002*gate_nosingularity(-(v-25),9 )
am   = lambda v: 0.182*gate_nosingularity(  v+35 ,9 )
bm   = lambda v: 0.124*gate_nosingularity(-(v+35),9 )
ah   = lambda v: 0.250*sexp(-(v+90)/12)
bh   = lambda v: 0.250*sexp(-(v+90)/12)*sexp((v+62)/6)
```

## Python code for integrating the model and plotting

```python
T  = 500 # ms total time
dt = 0.1 # ms integration step
N  = int(T/dt)
tt = arange(N+1)*dt

# Reversal potentials also from Table 2.1 in Gerstner et al. (2004)
v0      = -70 # Initial voltage
EN      =  55 # Sodium (Na⁺) reversal potential (mV)
EK      = -77 # Potassium (K⁺) reversal potential (mV)
EL      = -65 # Leak reversal potential (mV)
C       = 1.0 # Membrane capacitence nanofarads
I_apply = 0.5 # Nanoamperes
tstart  = T/4
tstop   = 3*T/4
I       = lambda t:I_apply if tstart<t<tstop else 0 # Time-varying input

def dstate_dt(t,state):
    v,h,n,m = state
    gK = gKmax*n**4
    gN = gNmax*m**3*h    
    IK = gK*(EK-v)
    IN = gN*(EN-v)
    IL = gL*(EL-v)
    dv = (IK + IN + IL + I(t))/C
    dn = (1-n)*an(v) - bn(v)*n
    dm = (1-m)*am(v) - bm(v)*m
    dh = (1-h)*ah(v) - bh(v)*h
    return float32([dv,dh,dn,dm])

# Integrates with built-in Runge–Kutta 45
start = time()
sol = solve_ivp(dstate_dt,(tt[0],tt[-1]),float32([v0,0,0,0]),t_eval=tt,method="RK45")
print("Elapsed: %6.4f s"%(time() - start))
state_rk45 = sol.y
hh_plot(tt, state_rk45, I_apply, tstart, tstop)
```

## Comparing integration schemes

- Normally you should use an appropriate **built-in solver** (fast, accurate), like `RK45` as demonstrated above. See the documentation of `solve_ivp` for more.
- **Forward Euler** is very easy to implement and debug, and might be used when code simplicity and interpretability is more important than accuracy. One place you might see Forward Euler is in gradient-descent learning rules that use a fixed time-step. You can view this form of leaning rule as integrating an ordinary differential equation to descend the gradient of the loss function.
- **Exponential Euler** almost as easy to implement and debug as forward Euler, but a bit slower. It is still first-order but uses assumptions that may provide a more appropriate bias for some problems. 

```python
# Forward Euler 
start = time()
state = zeros((N+1,4),'f') # v h n m
state[0,0] = v0
for i in range(N):
    state[i+1] = state[i] + dt*dstate_dt(i*dt,state[i])
print("Forward Euler took %6.4f s"%(time() - start))
state_fe = state.copy().T

# Forward Exponential Euler 
# Updates for gating variables, voltage are interleaved
# This is akin to a "leapfrog" solver and competes with rk45
start = time()
state = zeros((N+1,4),'f') # v h n m
state[0,0] = v0
def gating_ee_step(ap,bp,p,v,dt):
    α,β = ap(v),bp(v)
    return expm1(-dt*(α+β))*(p - α/(α+β))
for i in range(N):
    v,h,n,m = state[i]
    # Update gating variables
    h += gating_ee_step(ah,bh,h,v,dt)
    n += gating_ee_step(an,bn,n,v,dt)
    m += gating_ee_step(am,bm,m,v,dt)
    # Update voltage
    gK,gN = gKmax*n**4, gNmax*m**3*h    
    g  = gK+gN+gL
    v_ = (gK*EK + gN*EN + gL*EL + I(i*dt))/g
    v += expm1(-dt/C*g)*(v - v_)
    state[i+1] = v,h,n,m
print("Exponential forward Euler took: %6.4f s"%(time() - start))
state_ee = state.copy().T

figure(0,(8,2),120)
plot(tt,state_rk45[0],label='RK45')
plot(tt,state_fe[0],label='Forward Euler Δt=%s'%dt)
plot(tt,state_ee[0],label='Forward Exponential Euler Δt=%s'%dt, linestyle=':')
simpleraxis()
base_legend(fudge=-.5)
xlabel("Time (s)")
ylabel("Voltage (mV)")
show()
```
